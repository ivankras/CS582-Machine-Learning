### Columns with null values

```python
# Get names of columns with missing values
cols_with_missing = [col for col in imdb_ratings.columns if imdb_ratings[col].isnull().any()]
print(cols_with_missing) # result: ['arrival_delay_in_minutes']
```

```python
print([len(imdb_ratings[col][imdb_ratings[col].isnull()]) for col in cols_with_missing]) # how many values are missing
print([len(imdb_ratings[col]) for col in cols_with_missing]) # how many values are missing
# print(len(imdb_movies['country'][imdb_movies['country'].isnull()])) # how many values are missing
```
